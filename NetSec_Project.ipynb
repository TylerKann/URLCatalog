```python
df = pd.read_csv(file_name)

df.drop('Num', 1, inplace=True)
X = df.iloc[:,:-1].values
y = df['Site Type']
depth = 3
#y = tf.one_hot(y, depth)  # output: [1 x 3]
```

```python
print(X.shape)
print(y.shape)

# = y.numpy()
```

```python

SVC_model = SVC()
logreg_model = LogisticRegression()
KNN_model5 = KNeighborsClassifier(n_neighbors=5)
KNN_model20 = KNeighborsClassifier(n_neighbors=20)
bernouli_model = BernoulliNB()
gauss_model = GaussianNB()
dectree_model = DecisionTreeClassifier()
rf_model =  RandomForestClassifier()
gboost_model = GradientBoostingClassifier()
lda_model = LinearDiscriminantAnalysis()
mlp_model = MLPClassifier((10,10,10))


l = [SVC_model, logreg_model, KNN_model5, KNN_model20, bernouli_model, gauss_model, dectree_model, rf_model, gboost_model, lda_model, mlp_model]
accuracies = []
# x_train, x_test, y_train, y_test = train_test_split(X, df_y, test_size = 0.2, random_state = 4 )
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=27)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)

import time
start_time = time.time()

for model in l:

    model.fit(X_train,y_train)
  
    results = model.predict(X_test)
    print(type(model).__name__)
    accuracies.append(accuracy_score(results,y_test))
    print("Accuracy: \n", accuracy_score(results, y_test))
    print("Confusion: \n", confusion_matrix(results, y_test))
    print("Classification: \n", classification_report(results, y_test))
    print("***********************************************************")

print("time = " + str(time.time() - start_time))
```

```python
### graph
import plotly.graph_objects as go
names = []
for model in l:
    names.append((type(model).__name__))
accuracies2 = accuracies
order = np.argsort(np.asarray(accuracies2))
# print(order)
# np.flip(order)

accuracies2 = [accuracies2[i] for i in order]
names = [names[i] for i in order]
print(names)

# l = l[order]
names[7] = "KNN20"
fig = go.Figure(go.Bar(
            x=accuracies2,
            y=names,
            orientation='h'))

fig.show()
```
